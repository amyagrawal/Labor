```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from dateutil.relativedelta import relativedelta
import seaborn as sns
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('ASOS.csv', index_col=0)
df.index.name=None
df.reset_index(inplace=True)
df.drop(df.index[129], inplace=True)
```

```python
start = datetime.datetime.strptime("2006-01-01", "%Y-%m-%d")
date_list = [start + relativedelta(months=x) for x in range(0,129)]
df['index'] =date_list
df.set_index(['index'], inplace=True)
df.index.name=None
```

```python
df.columns= ['USA_LF', 'USA_UR', 'GSO_LF','GSO_UR','CLT_LF', 'CLT_UR', 'RDU_LF',
             'RDU_UR', 'Harrisburg_LF', 'Harrisburg_UR', 'RockHill_LF', 'RockHill_UR', 'Statesville_LF', 'Statesville_UR']
```

```python
df.RockHill_LF.plot(figsize=(12,8), title= 'RockHill Labor Force', fontsize=14).get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
#plt.savefig('month_ridership.png', bbox_inches='tight')
```

```python
decomposition_RockHill = seasonal_decompose(df.RockHill_LF, freq=12)  
fig = plt.figure()  
fig = decomposition_RockHill.plot()  
fig.set_size_inches(15, 8)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(df.RockHill_LF)
```

```python
df['RockHill_first_difference'] = df.RockHill_LF - df.RockHill_LF.shift(1)  
test_stationarity(df.RockHill_first_difference.dropna(inplace=False))
```

```python
df['RockHill_seasonal_difference'] = df.RockHill_LF - df.RockHill_LF.shift(12)  
test_stationarity(df.RockHill_seasonal_difference.dropna(inplace=False))
```

```python
df['RockHill_seasonal_first_difference'] = df.RockHill_first_difference - df.RockHill_first_difference.shift(12)  
test_stationarity(df.RockHill_seasonal_first_difference.dropna(inplace=False))
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df.RockHill_seasonal_first_difference.iloc[13:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df.RockHill_seasonal_first_difference.iloc[13:], lags=40, ax=ax2)
```

```python
mod = sm.tsa.statespace.SARIMAX(df.RockHill_LF, trend='n', order=(0,1,0), seasonal_order=(0,1,1,12))
results = mod.fit()
print results.summary()
```

```python
mod = sm.tsa.statespace.SARIMAX(df.RockHill_LF, trend='n', order=(0,1,0), seasonal_order=(1,1,1,12))
results = mod.fit()
print results.summary()
```

```python
df['RockHill_forecast'] = results.predict(start = 118, end= 130, dynamic= True)  
df[['RockHill_LF', 'RockHill_forecast']].plot(figsize=(12, 8)) 
#plt.savefig('ts_df_predict.png', bbox_inches='tight')
```
