```python
#Importing Libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from dateutil.relativedelta import relativedelta
import seaborn as sns
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#Reading the file
df = pd.read_csv('ASOS.csv', index_col=0)
df.index.name=None
df.reset_index(inplace=True)
df.drop(df.index[129], inplace=True)
```

```python
start = datetime.datetime.strptime("2006-01-01", "%Y-%m-%d")
date_list = [start + relativedelta(months=x) for x in range(0,129)]
df['index'] =date_list
df.set_index(['index'], inplace=True)
df.index.name=None
```

```python
df.columns= ['GSO_LF','GSO_UR','CLT_LF', 'CLT_UR', 'RDU_LF',
             'RDU_UR', 'Harrisburg_LF', 'Harrisburg_UR', 'USA_LF',
             'USA_UR']
```

```python
#Visualize the data
df.RDU_UR.plot(figsize=(12,8), title= 'RDU Unemployment Rate', fontsize=14)
#plt.savefig('month_ridership.png', bbox_inches='tight')
```

```python
decomposition_RDU = seasonal_decompose(df.GSO_UR, freq=12)  
fig = plt.figure()  
fig = decomposition_RDU.plot()  
fig.set_size_inches(15, 8)
```

```python
#Visualizing the time series and testing for stationarity
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(df.RDU_UR)
```

```python
#Taking a first difference of the data which should help to eliminate the overall trend from the data
df['RDU_first_difference'] = df.RDU_UR - df.RDU_UR.shift(1)  
test_stationarity(df.RDU_first_difference.dropna(inplace=False))
```

```python
#While this helped to improve the stationarity of the data it is not there yet. So the next step is to take a seasonal difference to remove the seasonality of the data and see how that impacts the stationarity of the data.
df['RDU_seasonal_difference'] = df.RDU_UR - df.RDU_UR.shift(12)  
test_stationarity(df.RDU_seasonal_difference.dropna(inplace=False))
```

```python
#Compared to the original data this is an improvement, but we are not there yet. The next step is to take a first difference of the seasonal difference.
df['RDU_seasonal_first_difference'] = df.RDU_first_difference - df.RDU_first_difference.shift(12)  
test_stationarity(df.RDU_seasonal_first_difference.dropna(inplace=False))
```

```python
#Plotting the ACF and PACF charts and find the optimal parameters
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df.RDU_seasonal_first_difference.iloc[13:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df.RDU_seasonal_first_difference.iloc[13:], lags=40, ax=ax2)
```

```python
#Building Model
mod = sm.tsa.statespace.SARIMAX(df.RDU_UR, trend='n', order=(0,1,0), seasonal_order=(0,1,1,12))
results = mod.fit()
print results.summary()
```

```python
mod = sm.tsa.statespace.SARIMAX(df.RDU_UR, trend='n', order=(0,1,0), seasonal_order=(1,1,1,12))
results = mod.fit()
print results.summary()
```

```python
#Now that we have a model built, we want to use it to make forecasts. First using the model to forecast for time periods that we already have data for, so we can understand how accurate are the forecasts.
df['RDU_forecast'] = results.predict(start = 118, end= 130, dynamic= True)  
df[['RDU_UR', 'RDU_forecast']].plot(figsize=(12, 8)) 
#plt.savefig('ts_df_predict.png', bbox_inches='tight')
```
